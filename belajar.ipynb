```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score,classification_report
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('gaya_belajar_dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
x = df.drop("Label", axis=1) 
y = df["Label"]
```

```python
y
```

```python
# Buat model Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Evaluasi dengan Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, x, y, cv=skf, scoring='accuracy')
```

```python
print("Akurasi tiap lipatan:", scores)
print(f"Rata-rata akurasi: {scores.mean():.2f}")
print(f"Standar deviasi: {scores.std():.2f}")
```

```python
# Latih ulang model untuk simpan
import joblib
model.fit(x, y)
joblib.dump(model, "random_forest_gaya_belajar.pkl")
```

```python
# Prediksi untuk evaluasi dasar
y_pred = model.predict(x)
print("\nClassification Report:")
print(classification_report(y, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y, y_pred))
```
